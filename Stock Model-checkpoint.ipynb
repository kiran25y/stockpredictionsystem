# Importing Modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

import yfinance as yf
from datetime import date, timedelta

# Setting environment to ignore future warnings
import warnings
warnings.simplefilter('ignore')
```

# Extracting Data

```python
df_sent = pd.read_csv("Sentiment_data.csv", parse_dates=["Date"])
df_sent.head()
```

```python
data = yf.download('AAPL',start=df_sent.Date.min(), end=df_sent.Date.max(), interval='1d').reset_index()
df = data.merge(df_sent, on="Date")
```

```python
df.head()
```

```python
df.set_index("Date")["Close"].plot(kind="line", figsize=(15, 6))
plt.show()
```

# Preparing Data

```python
df.isnull().sum().sum()
```

```python
df.Close = df.Close.shift(periods=-1)
df.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
df.dropna(inplace=True)
```

```python
df_train = df[df.Date <= pd.to_datetime("2015-06-30 00:00:00")]
df_test = df[df.Date > pd.to_datetime("2015-06-30 00:00:00")]
```

```python
X_train = df_train.drop(["Date", "Close"], axis=1)
y_train = df_train.Close

X_test = df_test.drop(["Date", "Close"], axis=1)
y_test = df_test.Close
```

# Building Model

```python
# Importing Models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor

# Importing evaluation modules
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# check the performance on diffrent regressor
models = []
models.append(('XGBRegressor', XGBRegressor()))
models.append(('LinearRegression', LinearRegression()))
models.append(('Random Forest', RandomForestRegressor()))
models.append(('Decision Tree', DecisionTreeRegressor()))

train_l = []
test_l = []
mae_l = []
rmse_l = []
r2_l = []

import time
i = 0
for name,model in models:
    i = i+1
    start_time = time.time()
    
    # Fitting model to the Training set
    clf = model
    clf.fit(X_train, y_train)
    
    # Scores of model
    train = model.score(X_train, y_train)
    test = model.score(X_test, y_test)
    
    train_l.append(train)
    test_l.append(test)
    
    # predict values
    predictions = clf.predict(X_test)
    # RMSE
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    rmse_l.append(rmse)
    # MAE
    mae = mean_absolute_error(y_test,predictions)
    mae_l.append(mae)
    # R2 score
    r2 = r2_score(y_test,predictions)
    r2_l.append(r2)



    print("+","="*100,"+")
    print('\033[1m' + f"\t\t\t{i}-For {name} The Performance result is: " + '\033[0m')
    print("+","="*100,"+")
    print('Training Score  : ', train)   
    print("-"*50)
    print('Testing Score   : ', test)   
    print("-"*50)
    print('Root mean squared error (RMSE) : ', rmse)   
    print("-"*50)
    print('Mean absolute error (MAE) : ', mae)
    print("-"*50)
    print('R2 score : ', r2)
    print("-"*50)
    
    plt.figure(figsize=(15, 6))
    sns.lineplot(range(len(y_test)), predictions, label="Prediction")
    sns.lineplot(range(len(y_test)), y_test, label="Actual")
    plt.show()

    
    print("\t\t\t\t\t\t\t-----------------------------------------------------------")
    print(f"\t\t\t\t\t\t\t Time for detection ({name}) : {round((time.time() - start_time), 3)} seconds...")
    print("\t\t\t\t\t\t\t-----------------------------------------------------------")
    print()
    
```

```python
comp = pd.DataFrame({"Models": dict(models).keys(), "Training Score": train_l, "Testing Score": test_l, "MAE": mae_l, "RMSE": rmse_l, "R2 Score": r2_l})
comp
```
